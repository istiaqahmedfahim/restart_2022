```python
iris = sns.load_dataset('iris')
correlation = iris.corr()
```

correlation of the data is shown here .Standard correlation varies from -1 to 1 .It shows us how a data of a feature is increasing or decreasing by the increasment or decreasment of another feature data. 

```python
correlation
```

```python
sns.heatmap(correlation,cmap='coolwarm',annot=True)
```

From above heatmap of we can understand that the correlation with petal-length with petal-width is very high, tends to 1. So they have a high correlation.

```python
sns.set(style="whitegrid",color_codes=True)
sns.lmplot(x='sepal_length',y='petal_length',hue='species',data= iris,palette='Set1',markers=["o","s","D"])
```

```python
sns.pairplot(iris,hue='species',markers=["o","s","D"])
sns.pairplot(iris,kind='reg',hue='species',markers=["o","s","D"])
```

```python
sns.boxplot(data=iris,orient='h')
```
